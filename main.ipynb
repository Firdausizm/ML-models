```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline # huruf besar cik ALAYY
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import root_mean_squared_error

df_clean = pd.read_csv('df_clean.csv')


# feature engineering
# ngurusin kolom pertama yang kek anjengg
df_clean['flat_or_house'] = df_clean['house_type'].str.extract(r"\b(Flat|House)\b", expand=False) # bakal error keknya

df_clean['avg_rent_per_locality'] = (df_clean.groupby('locality')
                                    ['rent'].transform('mean')
                                    )
df_clean['avg_rent_per_city'] = (df_clean.groupby('city')
                                ['rent'].transform('mean')
                                )
# mengukur seberapa premium locality
df_clean['locality_premium_rate'] = df_clean.avg_rent_per_locality / df_clean.avg_rent_per_city




x = df_clean[['locality', 'city', 'area', 'beds', 'bathrooms', 'balconies', 'furnishing', 'area_rate', 'flat_or_house', 'avg_rent_per_locality', 'avg_rent_per_city', 'locality_premium_rate']]
y = df_clean['rent']
```

```python
# fase pre prosesing

# pisah data dulu
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=0)

# isi pipeline
kategori_onehot = ['city'] 
kategori_ordinal = ['furnishing']

preprocesor = ColumnTransformer(
   transformers=[
        ('kategori_city', OneHotEncoder(handle_unknown='ignore'), kategori_onehot),
        ('kategori_furn', OrdinalEncoder(), kategori_ordinal)
   ])


# model
model_1 = RandomForestRegressor(random_state=0)

print('fase berhasil')
```

```python
df_clean.head()          # SETELAH SELESAI  BUANG AJA CELL INI NGGA PENTING
konzz = df_clean['flat_or_house'].isnull().sum()
print(konzz)
```

```python
# ini pipelinenya

pipeline = Pipeline(steps = [
   ('preprocesor', preprocesor),
   ('model', model_1)
])


pipeline.fit(x_train, y_train)

print('pipeline aman')
```

```python
hasil_prediksi = pipeline.predict(x_test)


# mean absolute eror
mae = mean_absolute_error(y_test, hasil_prediksi)
print(f"Nilai MAE Model : {mae}")

# root mean squared error (MSE)
rmse = root_mean_squared_error(y_test, hasil_prediksi)
print(f"Nilai MSE Model : {rmse}")
```
