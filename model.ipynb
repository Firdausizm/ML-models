```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline # huruf besar cik ALAYY
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import root_mean_squared_error

df_clean = pd.read_csv('df_clean.csv')

x = df_clean[['locality', 'city', 'area', 'beds', 'bathrooms', 'balconies', 'furnishing', 'area_rate']]
y = df_clean['rent']
```

```python
# fase pre prosesing

# pisah data dulu
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=0)

# isi pipeline
kategori_onehot = ['city'] 
kategori_ordinal = ['furnishing']

preprocesor = ColumnTransformer(
   transformers=[
        ('kategori_city', OneHotEncoder(handle_unknown='ignore'), kategori_onehot),
        ('kategori_furn', OrdinalEncoder(), kategori_ordinal)
   ])

# model
model_1 = RandomForestRegressor(random_state=0)

print('fase berhasil')
```

```python
# ini pipelinenya

pipeline = Pipeline(steps = [
   ('preprocesor', preprocesor),
   ('model', model_1)
])


pipeline.fit(x_train, y_train)

print('pipeline aman')
```

```python
# ngampung bentar
hasil_prediksi = pipeline.predict(x_test)


# mean absolute eror      benatr jangan di run duluuu
mae = mean_absolute_error(y_test, hasil_prediksi)
print(f"Nilai MAE Model : {mae}")

# root mean squared error
rmse = root_mean_squared_error(y_test, hasil_prediksi)
print(f"Nilai RMSE Model : {rmse}")
```
